### COPYING ARRAYS

```python
#Shallow copy | Changes to y_array will affect original array
y_array = temp_arr
```

```python
temp_arr
```

```python
y_array[0] = [3,3,3]
y_array
```

```python
#Deep copy | Changes to new_arr will not affect original array
new_arr = np.copy(temp_arr)
new_arr
```
